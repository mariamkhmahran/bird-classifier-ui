## Data Resoultion Analyser

Before preparing your data and training yout model you need to check if the resoultion of your images are similar to that of the model. Why do you think this is important? 

```python
import os
import seaborn as sns
import numpy as np
from matplotlib.pyplot import imread
```

```python
dim1 = []
dim2 = []
for image_filename in os.listdir('./images'):
    name, ext = os.path.splitext(image_filename)
    try:
        if not ext == '.xml':
            img = imread('./images/'+image_filename)
            d1,d2,color = img.shape
            dim1.append(d1)
            dim2.append(d2)
            
    except: 
        print(image_filename)
        continue
```

```python
sns.jointplot(dim1,dim2)
```

```python
np.mean(dim1)
```

```python
print('Min:', np.min(dim1))
print('Max:', np.max(dim1))
```

```python
np.mean(dim2)
```

```python
print('Min:', np.min(dim2))
print('Max:', np.max(dim2))
```
